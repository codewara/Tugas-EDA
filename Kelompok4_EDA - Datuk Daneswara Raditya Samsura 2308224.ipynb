### Group Similar Datasets Records

Menggabungkan data pada dataset berdasarkan kabupaten/kota dan tahun

```python
konseling = konseling.groupby(['bps_nama_kabupaten_kota', 'tahun'])['jumlah_ibu_hamil_konseling_gizi'].sum().reset_index()
konseling = konseling[['bps_nama_kabupaten_kota', 'jumlah_ibu_hamil_konseling_gizi', 'tahun']]

kronis    = kronis.groupby(['bps_nama_kabupaten_kota', 'tahun'])['jumlah_ibu_hamil_kek_kunjungan'].sum().reset_index()
kronis    = kronis[['bps_nama_kabupaten_kota', 'jumlah_ibu_hamil_kek_kunjungan', 'tahun']]

highRisk  = highRisk.groupby(['bps_nama_kabupaten_kota', 'tahun'])['jumlah_ibu_hamil_resti'].sum().reset_index()
highRisk  = highRisk[['bps_nama_kabupaten_kota', 'jumlah_ibu_hamil_resti', 'tahun']]
```

### Q1: Seberapa besar peluang kelahiran bayi hidup di Jawa Barat?

```python
import matplotlib.pyplot as plt

# Filter dataset di mana status bayi lahir tidak hidup
deaths = statBayi[statBayi['status'] == 'MATI']

# Group data berdasarkan tahun total jumlah
total = deaths.groupby('tahun')['jumlah'].sum()

# visualisasi data
plt.figure(figsize=(10, 5))
plt.plot(total.index, total.values)
plt.xlabel('Tahun')
plt.ylabel('Jumlah Kelahiran Bayi Mati')
plt.title('Jumlah Kelahiran Bayi Mati di Jawa Barat per Tahun')
plt.show()

# Group data berdasarkan tahun dan hitung persentase kematian bayi per tahun
percentage = deaths.groupby('tahun')['jumlah'].sum() / statBayi.groupby('tahun')['jumlah'].sum() * 100

# point year where the plot is maximum and minumum with red and blue
max_year = percentage.idxmax()
min_year = percentage.idxmin()

# visualisasi data
plt.figure(figsize=(10, 5))
plt.plot(percentage.index, percentage.values)
plt.plot(max_year, percentage.max(), 'ro')
plt.plot(min_year, percentage.min(), 'bo')
plt.xlabel('Tahun')
plt.ylabel('Persentase Kelahiran Bayi Mati')
plt.title('Persentase Kelahiran Bayi Mati per Tahun')
plt.show()
```

#### Analysis

- Peluang kelahiran bayi hidup di Jawa Barat bervariasi antara 99.66% hingga 99.76% selama periode 2015 hingga 2023.
- Pada tahun 2015, persentase kelahiran bayi mati mencapai sekitar 0.34%, sehingga peluang kelahiran bayi hidup pada tahun tersebut adalah 99.66%.
- Pada tahun-tahun berikutnya (2016-2020), persentase kelahiran bayi mati menurun hingga mencapai titik terendah pada sekitar 0.24%, sehingga peluang kelahiran bayi hidup meningkat menjadi sekitar 99.76%.
- Pada tahun 2023, persentase kelahiran bayi mati naik kembali menjadi sekitar 0.32%, sehingga peluang kelahiran bayi hidup menjadi 99.68%.
- Tren keseluruhan menunjukkan adanya fluktuasi dalam peluang kelahiran bayi hidup, namun secara umum berada di kisaran 99.66% hingga 99.76%, dengan peluang terbaik di tahun-tahun antara 2017-2020.
Peluang kelahiran bayi hidup di tahun 2023 mengalami sedikit penurunan dibandingkan beberapa tahun sebelumnya (2020-2022), namun masih lebih baik dibandingkan dengan 2015.


### Q2: Berapa persentase kesiapan keluarga menghadapi program kehamilan dan pasca-kehamilan di Jawa Barat?

```python
# prompt: barplot pregVisit group by year and jenis_kunjungan

import matplotlib.pyplot as plt
import seaborn as sns

# Group data by year and jenis_kunjungan, then calculate the sum of jumlah
grouped_data = pregVisit.groupby(['tahun', 'jenis'])['jumlah'].sum().reset_index()

# Create a barplot
plt.figure(figsize=(10, 5))
sns.barplot(x='tahun', y='jumlah', hue='jenis', data=grouped_data)
plt.xlabel('Tahun')
plt.ylabel('Jumlah Kunjungan')
plt.title('Jumlah Kunjungan Ibu Hamil per Tahun dan Jenis Kunjungan')
plt.legend(title='Jenis Kunjungan')
plt.show()
```

Analysis

### Q3: Seberapa baik kualitas posyandu di Jawa Barat?

```python
# prompt: plot pembiayaan group by tahun and sumber_dana, y axis is sumber_dana counted

import matplotlib.pyplot as plt
# Group data by year and sumber_dana, then calculate the count of sumber_dana
grouped_data = biaya.groupby(['tahun', 'sumber_dana'])['sumber_dana'].count().reset_index(name='count')

# Create a barplot
plt.figure(figsize=(10, 5))
sns.barplot(x='tahun', y='count', hue='sumber_dana', data=grouped_data)
plt.xlabel('Tahun')
plt.ylabel('Jumlah Sumber Dana')
plt.title('Jumlah Sumber Dana Pembiayaan Kesehatan Ibu dan Anak per Tahun')
plt.legend(title='Sumber Dana')
plt.show()
```

```python
# prompt: barplot posyandu aktif

import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
sns.barplot(x='tahun', y='jumlah', data=posyandu)
plt.xlabel('Tahun')
plt.ylabel('Jumlah Posyandu Aktif')
plt.title('Jumlah Posyandu Aktif di Jawa Barat per Tahun')
plt.show()
```

#### Analysis

* Dari tahun 2019 hingga 2021, APBDES mendominasi sumber pendanaan untuk posyandu, dengan kontribusi jauh lebih besar dibandingkan sumber dana lainnya seperti iuran bulanan masyarakat.
* Ada peningkatan signifikan dalam kontribusi iuran bulanan masyarakat pada 2020 dan 2021, yang menunjukkan adanya dukungan masyarakat yang lebih besar terhadap posyandu. Meskipun tidak sebesar dana dari APBDES, kontribusi ini membantu memperkuat stabilitas pendanaan posyandu di masa sulit.
* Pada tahun 2020, jumlah posyandu aktif meningkat signifikan dibandingkan tahun 2019, dari sekitar 1000 menjadi lebih dari 2000 posyandu.
* Jumlah posyandu aktif kembali menurun pada tahun 2021, meskipun tetap lebih tinggi dibandingkan 2019. Ini menandakan bahwa kualitas layanan posyandu menurun, meski masih dalam batas yang cukup baik.
* Meskipun dana dari iuran masyarakat meningkat pada 2020 dan tetap stabil pada 2021, jumlah posyandu aktif tetap menurun, menunjukkan bahwa sumber dana yang cukup saja tidak menjamin stabilitas operasional posyandu.

```python
# prompt: merge plot anemia, kronis, high risk in one plot and make year integer not decimal, add anemia trend line until 2021

import numpy as np
import matplotlib.pyplot as plt

# Convert 'tahun' to integer
anemia['tahun'] = anemia['tahun'].astype(int)
kronis['tahun'] = kronis['tahun'].astype(int)
highRisk['tahun'] = highRisk['tahun'].astype(int)

# Group data by year and calculate the sum of jumlah for each dataset
anemia_grouped = anemia.groupby('tahun')['jumlah'].sum().reset_index()
kronis_grouped = kronis.groupby('tahun')['jumlah'].sum().reset_index()
highRisk_grouped = highRisk.groupby('tahun')['jumlah'].sum().reset_index()

# predict 2021 anemia value

# Create the plot
plt.figure(figsize=(10, 5))

plt.plot(anemia_grouped['tahun'], anemia_grouped['jumlah'], label='Anemia', marker='o')
plt.plot(kronis_grouped['tahun'], kronis_grouped['jumlah'], label='Kronis', marker='o')
plt.plot(highRisk_grouped['tahun'], highRisk_grouped['jumlah'], label='High Risk', marker='o')

# Add trend line for anemia until 2021
anemia_trend_data = anemia_grouped[anemia_grouped['tahun'] <= 2021]
z = np.polyfit(anemia_trend_data['tahun'], anemia_trend_data['jumlah'], 1)
p = np.poly1d(z)
plt.plot(anemia_trend_data['tahun'], p(anemia_trend_data['tahun']), linestyle='--', color='blue', label='Trend Anemia')


plt.xlabel('Tahun')
plt.ylabel('Jumlah Kasus')
plt.title('Tren Anemia, Kronis, dan High Risk di Jawa Barat')
plt.legend(loc='lower left')
plt.grid(True)
plt.show()
```
