Dalam data yang ditampilkan, terlihat bahwa untuk Kota Banjar pada tahun 2018, Kabupaten Sukabumi pada tahun 2020, dan Kabupaten Cianjur pada tahun 2020, tidak ada catatan kelahiran baik bayi laki-laki maupun bayi perempuan, baik yang hidup maupun yang mati. Hal ini mungkin disebabkan oleh kurangnya data atau tidak terdatanya kelahiran pada tahun tersebut.

Sebagai solusi, kami mengambil rata-rata kelahiran bayi yang hidup dan mati dari tahun-tahun lain bedasarkan bayi hidup dan bayi mati.

```python
def meanLH(data, tahun, daerah):
    return data[(data['tahun'] != tahun) &
                (data['daerah'] == daerah) &
                (data['status'] == 'HIDUP') &
                (data['jenis_kelamin'] == 'LAKI-LAKI')]['jumlah'].mean().astype('int64')

def meanPH(data, tahun, daerah):
    return data[(data['tahun'] != tahun) &
                (data['daerah'] == daerah) &
                (data['status'] == 'HIDUP') &
                (data['jenis_kelamin'] == 'PEREMPUAN')]['jumlah'].mean().astype('int64')
def meanLM(data, tahun, daerah):
    return data[(data['tahun'] != tahun) &
                (data['daerah'] == daerah) &
                (data['status'] == 'MATI') &
                (data['jenis_kelamin'] == 'LAKI-LAKI')]['jumlah'].mean().astype('int64')
def meanPM(data, tahun, daerah):
    return data[(data['tahun'] != tahun) &
                (data['daerah'] == daerah) &
                (data['status'] == 'MATI') &
                (data['jenis_kelamin'] == 'PEREMPUAN')]['jumlah'].mean().astype('int64')

def fillVal(data):
  for tahun in data['tahun'].unique():
      for daerah in data['daerah'].unique():
        subset = data[(data['tahun'] == tahun) & (data['daerah'] == daerah)]
        if len(subset) == 4 and (subset['jumlah'] == 0).all():

          # data pengganti untuk bayi laki-laki hidup
          data.loc[(data['tahun'] == tahun) &
                   (data['daerah'] == daerah) &
                   (data['status'] == 'HIDUP') &
                   (data['jenis_kelamin'] == 'LAKI-LAKI'), 'jumlah'] = meanLH(data, tahun, daerah)

          # data pengganti untuk bayi perempuan hidup
          data.loc[(data['tahun'] == tahun) &
                   (data['daerah'] == daerah) &
                   (data['status'] == 'HIDUP') &
                   (data['jenis_kelamin'] == 'PEREMPUAN'), 'jumlah'] = meanPH(data, tahun, daerah)

          # data pengganti untuk bayi laki-laki mati
          data.loc[(data['tahun'] == tahun) &
                   (data['daerah'] == daerah) &
                   (data['status'] == 'MATI') &
                   (data['jenis_kelamin'] == 'LAKI-LAKI'), 'jumlah'] = meanLM(data, tahun, daerah)

          # data pengganti untuk bayi perempuan mati
          data.loc[(data['tahun'] == tahun) &
                   (data['daerah'] == daerah) &
                   (data['status'] == 'HIDUP') &
                   (data['jenis_kelamin'] == 'PEREMPUAN'), 'jumlah'] = meanPM(data, tahun, daerah)
  return data

statBayi = fillVal(statBayi)
```

Tidak adanya posyandu aktif di KABUPATEN SUKABUMI dan KOTA CIREBON pada tahun 2019 kemungkinan besar disebabkan oleh kesalahan dalam pelaporan atau pencatatan data.
```python
def meanVal(data, tahun, daerah):
  return data[(data['tahun'] != tahun) &
              (data['daerah'] == daerah)]['jumlah'].mean().astype('int64')

def fillVal(data):
  for tahun in data['tahun'].unique():
    for daerah in data['daerah'].unique():
      subset = data[(data['tahun'] == tahun) & (data['daerah'] == daerah)]
      if (subset['jumlah'] == 0).all():
        data.loc[(data['tahun'] == tahun) &
                 (data['daerah'] == daerah), 'jumlah'] = meanVal(data, tahun, daerah)
  return data

posyandu = fillVal(posyandu)
```

## Data Visual Analyzation

### Q1: Seberapa besar peluang kelahiran bayi hidup di Jawa Barat?

```python
# filter dataset di mana status bayi lahir tidak hidup
deaths = statBayi[statBayi['status'] == 'MATI']

# grup data jumlah dan persentase berdasarkan tahun
total = deaths.groupby('tahun')['jumlah'].sum()
percentage = deaths.groupby('tahun')['jumlah'].sum() / statBayi.groupby('tahun')['jumlah'].sum() * 100

# tentukan poin poin ekstrim
pmax = percentage.idxmax()
pmin = percentage.idxmin()
vmin = total.idxmin()

# visualisasi data
plt.figure(figsize=(15, 5))
sns.lineplot(x=percentage.index, y=percentage.values, marker='o')
plt.plot(pmax, percentage.max(), 'ro')
plt.plot(pmin, percentage.min(), 'bo')
plt.plot(vmin, percentage[vmin], 'go')

# berikan keterangan tiap titik
for x, y, val in zip(percentage.index, percentage.values, total.values):
    label = f"{val} ({y:.3f}%)"
    plt.annotate(label, (x, y), textcoords="offset points", xytext=(0, 15), ha='center',
                 bbox=dict(facecolor='white', alpha=1, edgecolor='gray'))

plt.xlabel('Tahun')
plt.ylabel('Persentase Kelahiran Bayi Mati')
plt.title('Persentase Kelahiran Bayi Mati di Jawa Barat per Tahun', fontsize=14, weight='bold')
plt.grid(True)
plt.show()
```

#### Observasi & Analisis

Observasi :
- Pada tahun 2015, terjadi puncak kematian bayi tertinggi dengan lebih dari 3.000 kasus, diikuti oleh penurunan tajam pada tahun 2016-2017, dengan angka kematian bayi yang stabil di sekitar 2.200 kasus.
- Tahun 2018 menunjukkan sedikit peningkatan, naik ke angka sekitar 2.400 kematian bayi, sebelum kembali mengalami penurunan hingga titik terendah pada tahun 2020 (~2.000 kasus).
- Selama pandemi 2020 hingga 2023, jumlah kematian bayi kembali meningkat secara signifikan dengan angka kematian melonjak di atas 3.000 kasus.
- Persentase kematian bayi menunjukkan tren yang serupa. Pada tahun 2015, persentase kematian bayi melebihi 0,34%, namun turun drastis menjadi 0,24% pada tahun 2016-2017. Persentase ini kemudian sedikit meningkat pada 2018-2019 sebelum kembali menurun pada tahun 2020. Namun, kenaikan signifikan terlihat pada tahun 2021 hingga 2023 dengan persentase mencapai sekitar 0,32%.

Analisis:

Penurunan tajam dalam angka kematian bayi dari tahun 2015 hingga 2017, mencerminkan keberhasilan Program Keluarga Indonesia Sehat, Kementerian Kesehatan telah melaksanakan berbagai program selama dua tahun terakhir. Seperti capaian di lingkup program Kesehatan Masyarakat (Kesmas) yang meliputi penurunan angka kematian Bayi dan angka kematian Ibu, perbaikan akses kesehatan ibu dan bayi serta program perawatan maternal yang lebih baik<sup>[[1]](https://kemkes.go.id/id/rilis-kesehatan/inilah-capaian-kinerja-kemenkes-ri-tahun-2015-2017)</sup>. Hasil analisis tentang hubungan status KEK ibu hamil dengan kejadian BBLR di wilayah tersebut menunjukkan bahwa ibu hamil yang mengalami KEK memiliki risiko 4,317 kali lebih tinggi untuk melahirkan bayi dengan BBLR dibandingkan ibu yang tidak mengalami KEK<sup>[[13]](https://scholar.ui.ac.id/en/publications/analisis-status-anemia-ibu-hamil-terhadap-kejadian-berat-badan-la)</sup>.

Rumah sakit di salah satu daerah dengan persentase BBLR rendah juga menyediakan layanan kelas hamil "Ibu Pintar," yang dirancang sebagai kelompok belajar bagi ibu hamil dengan usia kehamilan antara 4 hingga 36 minggu (menjelang persalinan). Tujuan dari layanan ini adalah untuk meningkatkan pengetahuan, mengubah sikap, dan perilaku ibu agar lebih memahami kehamilan, perubahan tubuh, keluhan selama kehamilan, perawatan kehamilan, proses persalinan, masa nifas, kontrasepsi, dan perawatan bayi baru lahir<sup>[[14]](https://rsmmbogor.com/rs-dr-h-marzoeki-mahdi-bogor-memiliki-layanan-kelas-hamil-ibu-pintar310Z1i)</sup>.

Selain itu layanan Posyandu yang komprehensif juga mendukung rendahnya persentase BBLR. Program tersebut mencakup suplementasi gizi mikro, pemeriksaan rutin bagi ibu hamil, dan pemberian tablet tambah darah, yang berperan penting dalam menjaga kesehatan ibu dan bayi<sup>[[15]](https://jdih.bekasikota.go.id/jdih/web/uploads/2021pw322337.pdf)</sup>.

```python
# gabungkan dataset jumlah ibu hamil dan kematian ibu hamil berdasarkan tahun dan daerahnya
pregDeath = pd.merge(momDeath, pregnants, on=['daerah', 'tahun'], how='left')
pregDeath.rename(columns={'jumlah_x': 'jumlah_kematian', 'jumlah_y': 'jumlah_hamil'}, inplace=True)

# hitung persentase kematian ibu hamil berdasarkan kolom jumlah kematian dan jumlah ibu hamil
pregDeath['persentase'] = (pregDeath['jumlah_kematian'] / pregDeath['jumlah_hamil']) * 100
pregDeath = pregDeath[['daerah', 'jumlah_hamil', 'jumlah_kematian', 'persentase', 'satuan', 'tahun']]

# grup data yang telah dinormalisasi berdasarkan tahun
total = pregDeath.groupby('tahun')['jumlah_kematian'].sum().reset_index()
grouped = pregDeath.groupby('tahun')['persentase'].mean().reset_index()

# tentukan poin poin ekstrim
pmax = grouped['persentase'].idxmax()
pmin = grouped['persentase'].idxmin()

# lineplot
plt.figure(figsize=(15, 5))
sns.lineplot(x='tahun', y='persentase', data=grouped, marker='o')
plt.plot(grouped.iloc[pmax]['tahun'], grouped['persentase'].max(), 'ro')
plt.plot(grouped.iloc[pmin]['tahun'], grouped['persentase'].min(), 'bo')

for x, y, val in zip(grouped['tahun'], grouped['persentase'], total['jumlah_kematian']):
    label = f"{val} ({y:.4f}%)"
    plt.annotate(label, (x, y), textcoords="offset points", xytext=(0, -17), ha='center',
                 bbox=dict(facecolor='white', alpha=1, edgecolor='gray'))

# tampilkan plot
plt.xlabel('Tahun')
plt.ylabel('Persentase')
plt.tick_params(axis='both', pad=10)
plt.title('Persentase Kematian Ibu Hamil di Jawa Barat per Tahun', fontsize=14, weight='bold')
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(integer=True))
plt.tight_layout()
plt.grid(True)
plt.show()
```

#### Observasi

**Grafik Kematian Ibu Hamil:**
- Persentase kematian diperoleh dari total kematian ibu hamil pada tahun tertentu dibagi dengan total ibu hamil pada tahun tersebut lalu dikalikan 100.
- Persentase kematian ibu hamil mengalami kenaikan dari 0.0226% (227 kasus) pada tahun 2016 menjadi 0.0243% (183 kasus) pada tahun 2017. Angka ini merupakan persentase kematian ibu hamil tertinggi dari tahun 2016 - 2019.
- Persentase kematian ibu hamil mengalami penurunan dari 0.0243% (183 kasus) pada tahun 2017 menjadi 0.0194% (157 kasus) pada tahun 2018.
- Persentase kematian ibu hamil mengalami penurunan meskipun tidak signifikan dari tahun 2018 (157 kasus) ke 2019 (128 kasus). Angka ini merupakan persentase kematian ibu hamil terendah dari tahun 2016 - 2019.
